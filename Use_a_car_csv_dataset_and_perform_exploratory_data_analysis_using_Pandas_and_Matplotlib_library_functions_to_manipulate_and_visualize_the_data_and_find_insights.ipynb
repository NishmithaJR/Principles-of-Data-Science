1) Import the libraries

```python
#import the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
```

2) Load the latest using pandas read_csv() function

```python
#Load the car.csv dataset
car=pd.read_csv('/content/drive/MyDrive/Principles of Data Science/cars.csv')
```

3) Display the head of the dataset using the head() function

```python
#Display the head of the dataset
car.head()
```

4) Display the bottom 5 rows from the dataset using tail() function

```python
#Display the bottom 5 rows
car.tail(5)
```

5) Print summary statistics of the dataset using the describe() function

```python
#Analytical summary of the dataset 
car.describe(include='all')
```

6) Plot a histogram for all the variables

```python
car.hist(figsize=(20,30))
```

7) Box plot to visualize the relationship between vehicle size and engine hp

```python
sns.boxplot(x="Size", y="CITY (kWh/100 km)",data=car)
```

8) Build a pair plot using seaborn library

```python
sns.pairplot(car)
```

9)Drop irrelevant columns from the dataset using drop() function

```python
car=car.drop(['YEAR','RATING'], axis=1)
car.head(5)
```

10) Use rename() function to rename the columns

```python
#Rename the column names
car=car.rename(columns={"Model":"Vehicle model","Size":"Vehicle size"})
car.head(5)
```

11) Print the total number of duplicate rows

```python
#Rows containing duplicate data 
duplicate_rows_car=car[car.duplicated()]
print("Number of duplicate rows:", duplicate_rows_car.shape)
```

```python
#Count the rows before removing the data 
car.count()
```

12) Remove the duplicate rows using the drop_duplicate() function

```python
#Drop the duplicates
car=car.drop_duplicates()
car.head()
```

13) Drop the missing values from the dataset

```python
#Find the null values
print(car.isnull().sum())
```

```python
#Drop the missing values
car=car.dropna()
car.count()
```

14) Plot a histogram to find the number of acr per brand

```python
#Ploting a histogarm for number if cars per brand
car.Make.value_counts().nlargest(40).plot(kind="bar",figsize=(10,5))
plt.title("Number of cars by Make")
plt.ylabel("Number of cars")
plt.xlabel("Make");
```

15) Draw a correlation plot between the variables

```python
#Find the relations between the variables
plt.figure(figsize=(20,10))
c=car.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```
