```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_table("/content/drive/MyDrive/Principles of Data Science/fruit_data_with_colors.txt")
```

```python
data
```

```python
x = data.iloc[:, [4, 3]].values
```

```python
disp = data.iloc[:, 3]  # Assuming 'disp' is the column index of the 'disp' variable in the dataset
hp = data.iloc[:, 4]  # Assuming 'hp' is the column index of the 'hp' variable in the dataset

plt.scatter(disp, hp, s=100, c='blue', label='mass vs width')

plt.xlabel('mass')
plt.ylabel('width')
plt.legend()
plt.title('Scatter plot: massvs width')
plt.show()
```

```python
#training the K-means model on a dataset
kmeans = KMeans(n_clusters=2, init='k-means++', random_state= 42)
y_predict= kmeans.fit_predict(x)
```

```python
#visulaizing the clusters
plt.scatter(x[y_predict == 0, 0], x[y_predict == 0, 1], s = 100, c = 'blue', label = 'Cluster 1') #for first cluster
plt.scatter(x[y_predict == 1, 0], x[y_predict == 1, 1], s = 100, c = 'green', label = 'Cluster 2') #for second cluster

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroid')
plt.title('Cluster of fruits')
plt.xlabel('mass')
plt.ylabel('width')
plt.legend()
plt.show()
```
